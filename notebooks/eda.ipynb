```python
messages = pd.read_csv('messages.csv')
```

```python
messages
```

```python
behavior_cols = ['is_complained', 'is_blocked','is_purchased','is_opened','is_clicked', 'is_unsubscribed', 'is_hard_bounced','is_soft_bounced']
```

```python
ununique_idx = messages[behavior_cols].apply(lambda x :x.map({'t':True, 'f':False})).sum(1) > 1
ununique_idx
```

```python
messages[ununique_idx][behavior_cols].head()
```

```python
msg_with_email = messages.dropna(subset=['email_provider'])[:1000000]
nonunique_emails_idx = msg_with_email.groupby('client_id')['email_provider'].agg('nunique') > 1
nonunique_emails_idx[nonunique_emails_idx.values]
```

```python
msg_with_provider = messages.dropna(subset=['stream'])[:]
nonunique_provider_idx = msg_with_provider.groupby('client_id')['stream'].agg('nunique') > 1
nonunique_provider_idx[nonunique_provider_idx.values]
```

```python
msg_with_platform = messages.dropna(subset=['platform'])[:10000]
nonunique_platform_idx = msg_with_platform.groupby('client_id')['platform'].agg('nunique') > 1
nonunique_platform_idx[nonunique_platform_idx.values]
```
